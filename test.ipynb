```python
df_load_case = list(df_stress_input["Load"].unique())
df_load_case
```

```python
df_berdasarkan_stress_point = df_stress_input[(df_stress_input["Section Position"]=="Pos-2")][["Load","Sig-xx(Summation) (kN/m²)"]]
pd.DataFrame(df_berdasarkan_stress_point)
```

```python
df_filter_tes = df_stress_input.loc[(df_stress_input["Load"]=='L_1_7_TP_WN_1')&(df_stress_input["Section Position"]=='Pos-1'),"Sig-xx(Summation) (kN/m²)"]
pd.DataFrame(df_filter_tes)
```

```python
import pandas as pd

# Misalkan df_berdasarkan_stress_point dan df_load_case sudah didefinisikan sebelumnya

new_data = {}

for load_case in df_load_case:
    df_filter = df_berdasarkan_stress_point[df_berdasarkan_stress_point["Load"] == load_case]["Sig-xx(Summation) (kN/m²)"]
    new_data[load_case] = df_filter.values  # Mengambil nilai sebagai array

df_result = pd.DataFrame(new_data)
df_result
```
